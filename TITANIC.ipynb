```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

SOMEONE'S SPENDING ON LUNCH OVER PASTWEEK

```python
days=[0,1,2,3,4,5,6]
money_spent=[10,12,12,10,14,22,24]
plt.plot(days,money_spent)
plt.figure(figsize=(9,6),dpi=100) 
plt.plot(days,money_spent,color='red') 
plt.xlabel('days')
plt.ylabel('money_spent')
plt.title('Spending')
plt.show()
```

WE COULD ADD A FRIENDS LUNCH SPENDING FOR COMPARISON.THIS CAN BE VERY USEFUL IF WE WANT TO COMPARE TWO DATASETS WITH THE SAME SCALE AND AXIS CATEGORIES

```python
# Days of the week:
days = [0, 1, 2, 3, 4, 5, 6]
# Your Money:
money_spent = [10, 12, 12, 10, 14, 22, 24]
# Your Friend's Money:
money_spent_2 = [11, 14, 15, 15, 22, 21, 12]
# Plot your money:
plt.plot(days, money_spent)
# Plot your friend's money:
plt.plot(days, money_spent_2)
# Display the result:
plt.show()
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Taronyufa/CsvReader/master/titanic.csv")
df
```

```python
df.head(5)
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='Sex',y='Survived',data=df,hue='Pclass')
plt.show()
```

IN THIS BARPLOT WE TALK ABOUT THREE FEATURES, PCLASS TELL US ABOUT PASSENGER CLASS

SCATTER PLOT

```python
x=np.arange(891)
plt.figure(figsize=(12,7))
sns.scatterplot(x=x,y='Age',data=df,hue='Survived')
plt.show()
```

THE ABOVE GRAPH IS PLOTTED IN SEABORN 

LINE PLOT

```python
plt.figure(figsize=(12,7))
plt.title("Fare plot")
plt.plot(sorted(df['Fare']),color='g')
plt.xlabel('Index')
plt.ylabel('Fare')
plt.show()
```

 BOX PLOT

```python
plt.figure(figsize=(12,7))
plt.title('Age vs Survived vs Sex')
sns.boxplot(x='Sex',y='Age',hue='Survived',data=df)
plt.show()
```

(outlayers) , center represents  based on mean
